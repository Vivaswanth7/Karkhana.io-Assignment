```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the parametric equations of a Mbius strip
def mobius(u, v, R=1):
    x = (R + v * np.cos(u / 2)) * np.cos(u)
    y = (R + v * np.cos(u / 2)) * np.sin(u)
    z = v * np.sin(u / 2)
    return x, y, z

# Step 2: Set parameters and create meshgrid
R = 1           # Radius
w = 1           # Width of the strip
n = 100         # Resolution
u_vals = np.linspace(0, 2 * np.pi, n)
v_vals = np.linspace(-w / 2, w / 2, n // 2)
u, v = np.meshgrid(u_vals, v_vals)

# Step 3: Evaluate surface points
x, y, z = mobius(u, v, R)

# Step 4: Plot the Mbius strip
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue', alpha=0.8, edgecolor='k')
ax.set_title("Mbius Strip")
plt.show()

# Step 5: Compute surface area
def compute_surface_area(x, y, z, u, v):
    du = u[0, 1] - u[0, 0]
    dv = v[1, 0] - v[0, 0]

    dx_du, dx_dv = np.gradient(x, du, dv)
    dy_du, dy_dv = np.gradient(y, du, dv)
    dz_du, dz_dv = np.gradient(z, du, dv)

    # Cross product of tangent vectors
    cross_x = dy_du * dz_dv - dz_du * dy_dv
    cross_y = dz_du * dx_dv - dx_du * dz_dv
    cross_z = dx_du * dy_dv - dy_du * dx_dv

    dA = np.sqrt(cross_x**2 + cross_y**2 + cross_z**2)
    return np.sum(dA) * du * dv

# Step 6: Compute edge length (numerical arc length)
def compute_edge_length(R, w, n):
    u_vals = np.linspace(0, 2 * np.pi, n)
    v_edge = w / 2
    u_array = u_vals
    v_array = np.full_like(u_vals, v_edge)
    x, y, z = mobius(u_array, v_array, R)
    dx = np.diff(x)
    dy = np.diff(y)
    dz = np.diff(z)
    length = np.sum(np.sqrt(dx**2 + dy**2 + dz**2))
    return length * 2  # Two edges: v = w/2

# Step 7: Output results
area = compute_surface_area(x, y, z, u, v)
edge_length = compute_edge_length(R, w, n)

print(f"Approximate Surface Area: {area:.4f}")
print(f"Approximate Edge Length: {edge_length:.4f}")
```
